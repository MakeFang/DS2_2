### Importing keras

```python
import numpy as np
from keras.layers import Dense, Dropout, Activation
from keras.optimizers import SGD
from sklearn.model_selection import train_test_split
from keras.layers import Input
from keras.models import Model
```

### using the functional model, build a linear regression

```python
def linear_regression(X, y, epochs = 1000, batch_size = 1):
    inputs = Input(shape = (X.shape[1],))
    predictions = Dense(1, activation='linear')(inputs)
    lin_reg_model = Model(inputs=inputs, outputs=predictions)
    lin_reg_model.compile(loss='mean_squared_error', optimizer='sgd', metrics=['accuracy'])
    lin_reg_model.fit(X, y, epochs = epochs, batch_size = batch_size)
    return lin_reg_model
```

## Now for logistic regression in Keras

```python
def logistic_regression(X, y, epochs = 1000, batch_size = 1):
    inputs = Input(shape = (X.shape[1],))
    predictions = Dense(1, activation='sigmoid')(inputs)
    log_reg_model = Model(inputs=inputs, outputs=predictions)
    log_reg_model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
    log_reg_model.fit(X, y, epochs = epochs, batch_size = batch_size)
    return log_reg_model
```

### testing against sklearn with a toy case

```python
from sklearn import linear_model

# General a toy dataset:s it's just a straight line with some Gaussian noise:
xmin, xmax = -5, 5
n_samples = 100
np.random.seed(0)
X = np.random.normal(size=n_samples)
y = (X > 0).astype(np.float)
X[X > 0] *= 4
X += .3 * np.random.normal(size=n_samples)

X = X[:, np.newaxis]

# Fit the classifier
clf = linear_model.LogisticRegression(C=1e5, solver='lbfgs')
clf.fit(X, y)
```

```python
clf.coef_
```

```python
clf.intercept_
```

```python
clf.score(X,y)
```

```python
clf_keras = logistic_regression(X, y.reshape(-1,1), epochs = 1000, batch_size = 1)
```

```python
clf_keras.summary()
```

```python
clf_keras.layers[1].get_weights()
```

```python
clf_keras.evaluate(X,y.reshape(-1,1))
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import show
```

```python
x_test = np.arange(-5,10,0.1)
```

```python
def logit_func(x):
    return 1/(1+np.exp(-x))
```

```python
y_sk = logit_func( clf.coef_ * x_test + clf.intercept_ )
```

```python
clf_keras_weights = clf_keras.layers[1].get_weights()
```

```python
clf_keras_weights[1][0]
```

```python
y_keras = logit_func( clf_keras_weights[0][0][0]* x_test + clf_keras_weights[1][0])
```
